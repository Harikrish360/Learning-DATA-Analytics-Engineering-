```python
#*args

def adding(*args):
    total = 0
    for num in args:
        total += num
    return total

print(adding(1,3,5,7,8))
```

```python
#Set is Unordered, using set we can do union, intersection operation, dont allow duplicates, set is mutable, but update is not possible we need to remove 
#the value and add

uber_cities = ["Chennai", "Bangalore", "Delhi","Chennai","Bangalore","Kerala"]
unique_values = set(uber_cities)
print(unique_values)

```

```python
uber_cities1 = {"Chennai", "Bangalore", "Delhi","Kerala"}
uber_cities2 = {"Bangalore", "Delhi","Kerala", "Hyderabad"}

print(uber_cities1.union(uber_cities2))
print(uber_cities1.intersection(uber_cities2))
print(uber_cities2.difference(uber_cities1))

```

```python
#Add & Remove

uber_cities1.add("Karur")
print(uber_cities1)

uber_cities1.remove("Chennai")
print(uber_cities1)
```

```python
#update. This is the possible way to add a value in set

a= {1,2,4,5}
print(a)

a.remove(2)
a.add(8)
print(a)
```

```python
#Discard remove value without rising an error
#This is Error because 9 isnt in the set
a= {1,2,4,5}
print(a)

a.remove(9)
print(a)
```

```python
#To resolve this we use discard
a= {1,2,4,5}
print(a)

a.discard(9)
print(a)
```
