# Students Mark Analysis

1. Average by Subject and Students
2. Highest and lowest per subject 
3. Overall class toppers
4. Pass count per subject 
5. which subject is difficult
6. Ranking students

```python
import numpy as np

np.random.seed(40)
marks = np.random.randint(10, 101, size= (20,5))
print(marks)
```

```python
Avg_by_subject = np.mean(marks, axis =0)
print("Average in subjects : ",Avg_by_subject)

Avg_by_students = np.mean(marks, axis= 1)
print("Average of students : ", Avg_by_students)
```

```python
Highest_mark = np.max(marks, axis=0)
print("Max mark per subjects : ",Highest_mark)

Lowest_mark = np.min(marks ,axis= 0)
print("Min mark per subject : ", Lowest_mark)
```

```python
#Overall topper

total_marks = np.sum(marks , axis =1) 
print("Total marks by students : ",total_marks)

topper = np.argmax(total_marks) #argmax pick the position of max value
print(f"Topper of the class is {topper} with overall mark of {total_marks[topper]}")
```

```python
# pass count

pass_fail = marks >= 40

pass_count = np.sum(pass_fail, axis= 0)
print("Count of pass: ", pass_count)
```

```python
#difficult subject

Avg_by_subject = np.mean(marks, axis=0)
print("Average marks in subjects", Avg_by_subject)

difficult_subject = np.argmin(Avg_by_subject)
print("Difficult subject is", difficult_subject, "with avg of mark of", Avg_by_subject[difficult_subject])
```

```python
#Ranks 

ranks = np.argsort(-total_marks) # - helps in arrange for descending order
print(f"First Rank of the class is , {ranks [0]}")
print(f"Top 5 of the class is, {ranks [:5]}")
```
