```python
# work on sales data

df=pd.read_csv('sales_data.csv', skipinitialspace=True)
print(df)
```

```python
# Add new column

df['Total']= df['Price'] * df['Quantity']

# Group by products

grouped = df.groupby("Product")['Total'].sum().reset_index() #reset -> helps to reset the column

# Sort in desc

sorted_df = grouped.sort_values(by= 'Total', ascending = False)

print("Sales Summary")
print(sorted_df)
```

```python
# Joins

customer = pd.DataFrame({
    'C_Id': [101,102,103],
    'Name': ['Alice','bob','charlie'],
})

orders = pd.DataFrame({
    'Order_Id': [1,2,3,4,5],
    'C_Id': [101,102,103,101,103],
    'Food' : ['Biriyani',"Parotta","Fried Rice", "Noodeles", "Pasta"]
})

result = pd.merge(customer,orders, on= 'C_Id', how='inner')
print("Inner Join")
print(result)
```
