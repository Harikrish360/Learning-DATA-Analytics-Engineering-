```python
#Override 

class dad:
    def house(self):
        print("This 3 bhk house in Blue Color")

class son(dad):
    def factory(Self):
        print("The factory is developed by me")
        
#override the house method
    def house(self):
        print("I change the color of dad house into royal Green")

S= son()
S.factory()
S.house()
```

```python
#Multilevel Inheritance -> class a() -> class b(a), class c(b)

class grandfather:
    def assets(self):
        print("Farm Land and house is grand father asset")

class father(grandfather): # father class acces  the grandfather class
    def house(self):
        print("House is father asset")

class son(father):
    def factory(self):
        print("Factory is my asset")

s = son() #using son we can access to both grandfather and father class
s.factory()
s.house()
s.assets()
```

```python
# hierarchical inheritance -> class a -> class b(a) ->class c(a)

class father(): #main class
    def house(self):
        print("House is father asset")

class son1(father):       #Sub class
    def factory(self):
        print("Factory is my asset")

class son2(father):      #Sub class
    def car(self):
        print("Car is my asset")

#In hierarchy we need to create different  objects for the class both son1, son2 have access to fathers property, but son1 cant access his brother property same follow for son2
s1= son1()
s1.house()
s1.factory()

s2= son2()
s2.house()
s2.car()


```

```python
#Hybrid (more than 1 type of inheritance used inside the class is Multiple) 
#Mostly we didnt use Hybrid

class Engine:
    def engine_info(self):
        print("This is a powerful engine.")

class Body:
    def body_info(self):
        print("This is a steel body.")

# Multiple Inheritance: Car inherits from Engine and Body
class Car(Engine, Body):
    def car_info(self):
        print("This is a standard car.")

# Hierarchical Inheritance: Both ElectricCar and PetrolCar inherit from Car
class ElectricCar(Car):
    def battery_info(self):
        print("This car runs on a 75kWh battery.")

class PetrolCar(Car):
    def fuel_info(self):
        print("This car runs on Octane fuel.")

# Usage
tesla = ElectricCar()
tesla.engine_info()  # From Engine
tesla.body_info()    # From Body
tesla.car_info()     # From Car
tesla.battery_info() # From ElectricCar
```

```python
from abc import ABC, abstractmethod

class manager(ABC):     #To access abstract method in class we need to mention (ABC) in Parent class, and never create a object for Abstract class directly.
                        #Always call the abstract class method by using child class only.
    @abstractmethod
    def login(self):            #This is abstract method, pass is for future implementation
        pass

    @abstractmethod             # This is abstract method
    def logout(self):            
        pass

    def checkout(self):          #This is normal method
        pass

class developer(manager):

#overriding
    def login(self):
        print("Login page is completed")

    def logout(self):
        print("Logout page is completed")

    def checkout(self):
        print("Checkout page is completed")

dev = developer()
dev.login()
dev.logout()
dev.checkout()
```

```python
#We can use access specifiers inside the parent class
class parent():

    def __init__(self):
        self.public = "Im Public"
        self.__private = "Im Private"
        self._protected = "Im Protected" 

    def acessinsideparent(self):
        print("I can access them")
        print("Public :", self.public )
        print("Private :", self.__private)
        print("Protected :", self._protected)

p = parent()
p.acessinsideparent()
```

```python
#We can use access specifier using child class also

class parent():

    def __init__(self):
        self.public = "Im Public"
        self.__private = "Im Private"
        self._protected = "Im Protected" 

    def acessinsideparent(self):
        print("I can access them")
        print("Public :", self.public )
        print("Private :", self.__private)
        print("Protected :", self._protected)

class child(parent):                 #In sub class we cant able to call the private variable.

    def accesswithchild(self):
        print("This is child class")
        print("Child Public :", self.public )
        print("Child Protected :", self._protected)

        print("Child Private :", self._parent__private)

        try:
            print("Child Private :", self.__private)  # to access the private from child or staranger class also possible using Name mangling.
                                                      #print("Child Private :", self._parent__private)
        except AttributeError:
            print("Private :  Cannot access (Attribute Error)")

c=child()
c.accesswithchild()
```

# Access specifier for method

```python
class parent():

    def public_method(self):
        print("This is public Method")

    def __private_method(self):
        print("This is Private Method")

    def _protected_method(self):
        print("This is Protected Method")

    def access_from_same_class(self):
        print("Inside Parent")
        self.public_method()
        self.__private_method()
        self._protected_method()

p = parent()
p.access_from_same_class()
```

```python
# With child call method

class child(parent):

    def access_from_sub_class(self):
        print("Inside Child class (Sub class)")
        self.public_method()
        self._protected_method()
        try:
            self.__private_method()
        except AttributeError:
            print("Private method is cannot accessible")

C=child()
C.access_from_sub_class()

    
```

```python
#In this problem we hide the dicount amnt calculation from user 

class order:

    def __init__(self, customer_name, item, total_amount, discount):
        self.customer_name = customer_name                             #public
        self.item = item                                               #public
        self.__total_amount = total_amount                             #private
        self.__discount = discount                                     #private


    def __calculate_final_amount(self):                                    #private helper method, by using this we hide the info of disount reduction / hide coupons from user
        return self.__total_amount - self.__discount                       #Hide the info  of calculation
    
    def _get_admin_view(self):                                             #Protected method for admin view
        return {
            "Customer Name": self.customer_name,                           #Dictionary used because 
            "Item": self.item,
            "Total_amount" : f" Rs{self.__total_amount}",
            "Discount": f"Rs{self.__discount}",                              #Hide the calculation info for admin
            "Final_bill": f"Rs{self.__calculate_final_amount()}"
        }

    def customer_view(self):                                                #Customer view is public and beloc dict key and value only visible to cutomers

        return {
            "Customer Name": self.customer_name,
            "Item": self.item,
            "Final_bill": f"Rs{self.__calculate_final_amount()}"
        }
    

#everything created in a same class, but calling them in a different way

class admin_portal:
    def show_orders(self, order):
        return order._get_admin_view()             #Allow access to admin to see, but protected
    
class customer_app:                                #Public view but cant see the calc 
    def show_orders(self,order):
        return order.customer_view()

order =order("Jame",["Pepsi", "Pizza", "wings"],900, 150)

admin = admin_portal()
customer = customer_app()

print("Admin View")
print(admin.show_orders(order))

print(" \n Customer App")
print(customer.show_orders(order))

```

```python
class Myclass:
    def Instance_method(self):
        print("Called Instance method")

    @classmethod                 #the symbo @declaration is called annotation, and this class method dont depend on objects, it is class level method.
    def Class_method(cls):
        print("called class method")

    @staticmethod                #static method in independent dont depend on cls or obj(self)
    def Static_method():
        print("Called Static method")

obj=Myclass()
obj.Instance_method()

Myclass.Class_method()
Myclass.Static_method()
```

```python
#Classmethod Example -> we can override class level data

class Employee: 
    company_name = "Open Ai"  #class level data, global variable 

    @classmethod 
    def change_company(cls, new_name): #new_name -> argument
        cls.company_name = new_name    #Accesing Class variable

Employee.change_company("Google")
print(Employee.company_name)
```

```python
#StaticMethod ->static cant override class or object level data. Using static we can run operations only inside the static method declared class.

class math:

    @staticmethod
    def add(X,Y):
        return X + Y
    
print(math.add( 6 , 6))
```

```python
#Example with Class and static

class employee:
    company_name = "Open Ai"

    @classmethod 
    def change_company(cls, new_name): 
        cls.company_name = new_name 

    @staticmethod
    def try_change_company(new_name):
        company_name = new_name

employee.change_company("Amazon")
print(employee.company_name)

employee.try_change_company("Meta") #it wont change to meta, instead show Amazon only
print(employee.company_name)
```

w+ -> write + read (over write or create)
a+ -> append + read (create if not exist)
rb -> read binary , wb -> write binary , ab -> append binary

```python
#Write

file = open("note.txt","w")
file.write("welcome to Python file handling! \n")
file.write("This is new file \n")
file.close()                                        #never forget to close the file once it get opens
```

```python
#Read

file = open("note.txt","r")
content = file.read()      # this line helps to store the all data from note.txt -> content
print("File Contents \n", content)
file.close()
```

```python
#Append
file = open("note.txt", "a") #instead a we place "w" means it overwrite the whole file
file.write("This is completely ne3 3rd line. \n")
file.close()
```

```python
#with -> is proper pythonic way of writing file handling, by using with we dont need to mention the close().
   
with open("note.txt", "r") as file: 
        for line in file:          #for loop in file handling work as a readline() method
              print(line.strip())
```

```python
with open("file.txt") as f:
    while True :
        line = f.readline()
        if not line:
            break
        if "ERROR" in line:                       #Finding "ERROR" pattern words in the file.txt
            print("Found Error :", line.strip())
```

```python
#Throwaway variable + Range limit

with open("file.txt") as f:
    for _ in range(3):        # "_" underscore is throw away variable, we set range as 3. 
        print(f.readline().strip())   #Only in readline() method we able to set the range limit
```

```python
# read the file and copy paste the data in another file

with open("Existing.csv", "r") as input_file , open("New_file.csv", "w") as output_file:
    for line in input_file:
        print(line.strip())
        output_file.write(line)
```

```python
#retrieve the age data from the csv file
 
import csv            #import csv library

with open("Existing.csv","r") as file:
    reader = csv.DictReader(file) # Dictreader work as a dictionary helps to fetch the header from our csv file
    for row in reader:
        print(row["age"])
```
